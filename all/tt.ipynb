```python
# webcam_capture.py
import cv2
import os

# Adjust paths
SAVE_FOLDER = r"C:\Users\USER\Documents\ATTNDB\facerecognition"
os.makedirs(SAVE_FOLDER, exist_ok=True)

name = input("Enter full name: ")
filename = f"{name}.jpg"
save_path = os.path.join(SAVE_FOLDER, filename)

cap = cv2.VideoCapture(0)
print("üì∏ Press 'c' to capture or 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Failed to read from webcam.")
        break

    cv2.imshow("Live Camera", frame)
    key = cv2.waitKey(1)

    if key & 0xFF == ord('c'):
        cv2.imwrite(save_path, frame)
        print(f"‚úÖ Saved image as {save_path}")
        break
    elif key & 0xFF == ord('q'):
        print("‚ùå Quit without saving.")
        break

cap.release()
cv2.destroyAllWindows()
# webcam_capture.py
import cv2
import os

# Adjust paths
SAVE_FOLDER = r"C:\Users\USER\Documents\ATTNDB\facerecognition"
os.makedirs(SAVE_FOLDER, exist_ok=True)

name = input("Enter full name: ")
filename = f"{name}.jpg"
save_path = os.path.join(SAVE_FOLDER, filename)

cap = cv2.VideoCapture(0)
print("üì∏ Press 'c' to capture or 'q' to quit.")

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Failed to read from webcam.")
        break

    cv2.imshow("Live Camera", frame)
    key = cv2.waitKey(1)

    if key & 0xFF == ord('c'):
        cv2.imwrite(save_path, frame)
        print(f"‚úÖ Saved image as {save_path}")
        break
    elif key & 0xFF == ord('q'):
        print("‚ùå Quit without saving.")
        break

cap.release()
cv2.destroyAllWindows()
```
